```
data = pd.read_csv('/content/drive/MyDrive/Data/ITC.NS.csv')
df_itc = pd.DataFrame(data)
df_itc.head()
```

```
'''
This is a function to draw a line graph of given two series values. The function will ask 6 arguments, which are-

xlabel,
ylabel,
title,
x-axis value of first one, 
x-axis value of second one, 
Y-axsis values of base one,
'''

def plotMyGraph(x_axis, y_axis_1,y_axis_2, x_label, y_lable, fig_title):
  plt.plot(x_axis, y_axis_1)
  plt.plot(x_axis, y_axis_2)

  plt.xlabel(x_label)
  plt.ylabel(y_lable)
  plt.title(fig_title)
  plt.show()
```

```
plotMyGraph(df_itc['Date'], df_itc['Open'], df_itc['Close'],  'Date', 'Stock Price', 'Open Price vs Closing Price')
```

```
plotMyGraph(df_itc['Date'], df_itc['High'], df_itc['Low'], 'Date', 'Stock Price', 'High Price vs Low Price')
```

```
plotMyGraph(df_itc['Date'], df_itc['High'], df_itc['Open'], 'Date', 'Stock Price', 'High Price vs Open Price')
```

```
plotMyGraph(df_itc['Date'], df_itc['Adj Close'], df_itc['Open'], 'Date', 'Stock Price', 'Adj Close Price vs Open Price')
```

```
#Average price per day
df_itc['Avg Per Day'] = (df_itc['High']+df_itc['Low'])/2
```

```
plotMyGraph(df_itc['Date'], df_itc['Avg Per Day'], df_itc['Open'], 'Date', 'Stock Price', 'Avg Per Day Price vs Open Price')
```

```
plt.plot(df_itc['Open'],df_itc['Volume'])

plt.xlabel("Open Price")
plt.ylabel("Volume")
plt.title("Relation between Open price in respective of Volume")
plt.show()
```

```
plt.plot(df_itc['Date'],df_itc['Volume'])

plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Relation between Date in respective of Volume")
plt.show()
```
